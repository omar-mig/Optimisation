```python
# --- Grid ---
Nx = 200
L  = 1000.0     # meters
dx = L / Nx
x  = np.linspace(dx/2, L-dx/2, Nx)

# --- Rock ---
phi = 0.20      # porosity (constant for baseline)
k   = 100e-15   # permeability in m^2 (â‰ˆ100 mD)

# --- Fluids (gas=H2, water=brine) ---
mu_g = 1.0e-5   # Pa.s (order of magnitude for H2)
mu_w = 8.0e-4   # Pa.s (brine-ish)

# --- Corey-type relperm params ---
Sgr = 0.05      # residual gas
Swr = 0.20      # residual water

ng = 2.0
nw = 2.0

def clampSg(Sg):
    return np.clip(Sg, Sgr, 1.0 - Swr)

def krg(Sg):
    Sg = clampSg(Sg)
    Se = (Sg - Sgr) / (1.0 - Swr - Sgr)
    return Se**ng

def krw(Sg):
    Sg = clampSg(Sg)
    Sw = 1.0 - Sg
    Se = (Sw - Swr) / (1.0 - Swr - Sgr)
    return Se**nw

def fractional_flow_g(Sg):
    lg = krg(Sg) / mu_g
    lw = krw(Sg) / mu_w
    return lg / (lg + lw)
```

```python
def step_transport(Sg, u, dt):
    """
    1D upwind scheme for: dS/dt + (u/phi) d f(S)/dx = 0
    """
    f = fractional_flow_g(Sg)
    flux = u * f  # volumetric flux of gas fraction

    # Upwind numerical flux
    F = np.zeros(Nx+1)
    if u >= 0:
        F[1:Nx] = flux[:-1]
        F[0]    = u * fractional_flow_g(1.0 - Swr)  # injected gas saturation state
        F[Nx]   = flux[-1]
    else:
        F[1:Nx] = flux[1:]
        F[0]    = flux[0]
        F[Nx]   = u * fractional_flow_g(Sgr)         # production boundary

    # Update
    dS = -(dt/(phi*dx)) * (F[1:] - F[:-1])
    Snew = clampSg(Sg + dS)
    return Snew
```

```python
# Initial condition: mostly brine (low gas)
Sg = np.ones(Nx) * Sgr

# Schedule (seconds)
day = 24*3600
schedule = [
    ("inject",   5*day,  +1.0e-5),  # u > 0 pushes gas in
    ("shutin",   2*day,   0.0),
    ("withdraw", 5*day,  -1.0e-5),  # u < 0 pulls gas back
]

dt = 0.2*day  # time step; reduce if unstable

history = []
time = 0.0

for (name, duration, u) in schedule:
    nsteps = int(np.ceil(duration/dt))
    for _ in range(nsteps):
        Sg = step_transport(Sg, u, dt)
        time += dt
        history.append((time, name, Sg.copy()))
```

```python
!pip -q install scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import lil_matrix, csr_matrix
from scipy.sparse.linalg import spsolve
```

```python
# ---------------------------
# Grid
# ---------------------------
Nx, Ny = 80, 80
Lx, Ly = 1000.0, 1000.0  # meters
dx, dy = Lx/Nx, Ly/Ny

x = (np.arange(Nx) + 0.5) * dx
y = (np.arange(Ny) + 0.5) * dy
X, Y = np.meshgrid(x, y, indexing="ij")

# ---------------------------
# Rock/fluid (baseline)
# ---------------------------
phi = 0.20
k   = 150e-15      # m^2 (â‰ˆ150 mD)
mu  = 1.0e-3       # Pa.s (use single-phase viscosity for pressure solve baseline)

# ---------------------------
# Relperm / fractional flow for transport
# ---------------------------
mu_g = 1.0e-5
mu_w = 8.0e-4
Sgr, Swr = 0.05, 0.20
ng, nw = 2.0, 2.0

def clampSg(Sg):
    return np.clip(Sg, Sgr, 1.0 - Swr)

def krg(Sg):
    Sg = clampSg(Sg)
    Se = (Sg - Sgr) / (1.0 - Swr - Sgr)
    return Se**ng

def krw(Sg):
    Sg = clampSg(Sg)
    Sw = 1.0 - Sg
    Se = (Sw - Swr) / (1.0 - Swr - Sgr)
    return Se**nw

def fg(Sg):
    lg = krg(Sg)/mu_g
    lw = krw(Sg)/mu_w
    return lg/(lg+lw)
```

```python
def ij_to_id(i, j):
    return i*Ny + j

def build_pressure_matrix(Nx, Ny, dx, dy, k, mu):
    N = Nx*Ny
    A = lil_matrix((N, N))
    T_x = (k/mu) / dx**2
    T_y = (k/mu) / dy**2

    for i in range(Nx):
        for j in range(Ny):
            pid = ij_to_id(i, j)
            diag = 0.0

            # West/East
            if i > 0:
                nid = ij_to_id(i-1, j)
                A[pid, nid] = -T_x
                diag += T_x
            if i < Nx-1:
                nid = ij_to_id(i+1, j)
                A[pid, nid] = -T_x
                diag += T_x

            # South/North
            if j > 0:
                nid = ij_to_id(i, j-1)
                A[pid, nid] = -T_y
                diag += T_y
            if j < Ny-1:
                nid = ij_to_id(i, j+1)
                A[pid, nid] = -T_y
                diag += T_y

            A[pid, pid] = diag

    # Fix gauge (pressure reference) at cell (0,0)
    ref = ij_to_id(0, 0)
    A[ref, :] = 0.0
    A[ref, ref] = 1.0
    return A.tocsr()

A = build_pressure_matrix(Nx, Ny, dx, dy, k, mu)

def solve_pressure(q):
    """
    q is Nx x Ny source term (positive injection, negative production).
    """
    b = q.reshape(-1).copy()
    b[ij_to_id(0, 0)] = 0.0  # reference pressure cell
    p = spsolve(A, b).reshape((Nx, Ny))
    return p
```

```python
inj = (int(0.25*Nx), int(0.50*Ny))
prd = (int(0.75*Nx), int(0.50*Ny))

print("Injector cell:", inj, "Producer cell:", prd)
```

```python
def velocities_from_pressure(p, k, mu, dx, dy):
    """
    Returns face velocities u_x (Nx+1, Ny), u_y (Nx, Ny+1).
    Positive u_x is +x direction; positive u_y is +y direction.
    """
    Tx = (k/mu)/dx
    Ty = (k/mu)/dy

    u_x = np.zeros((Nx+1, Ny))
    u_y = np.zeros((Nx, Ny+1))

    # Internal faces: u = - (k/mu) dp/dx
    # x-faces between i-1 and i: use p[i]-p[i-1]
    for i in range(1, Nx):
        u_x[i, :] = -Tx * (p[i, :] - p[i-1, :])

    for j in range(1, Ny):
        u_y[:, j] = -Ty * (p[:, j] - p[:, j-1])

    # No-flow boundaries already enforced by leaving boundary faces = 0
    return u_x, u_y
```

```python
def transport_step(Sg, u_x, u_y, dt, phi, dx, dy, phase, inj, prd, Swr, Sgr):
    """
    Conservative upwind transport:
      dS/dt + div( u * f(S) ) / phi = 0

    Added well coupling:
    - During injection: enforce high gas saturation in injector cell (Dirichlet).
    - Optionally during withdrawal: enforce low gas saturation at producer to mimic removal.
    """
    Sg = clampSg(Sg)
    f = fg(Sg)

    Fx = np.zeros((Nx+1, Ny))
    Fy = np.zeros((Nx, Ny+1))

    # x-faces (upwind)
    for i in range(1, Nx):
        ux = u_x[i, :]
        up = (ux >= 0)
        fL = f[i-1, :]
        fR = f[i, :]
        Fx[i, :] = ux * np.where(up, fL, fR)

    # y-faces (upwind)
    for j in range(1, Ny):
        uy = u_y[:, j]
        up = (uy >= 0)
        fB = f[:, j-1]
        fT = f[:, j]
        Fy[:, j] = uy * np.where(up, fB, fT)

    divF = (Fx[1:, :] - Fx[:-1, :]) / dx + (Fy[:, 1:] - Fy[:, :-1]) / dy
    Snew = clampSg(Sg - (dt/phi) * divF)

    # -----------------------------
    # WELL COUPLING (THE FIX)
    # -----------------------------
    if phase == "inject":
        # Force injector cell to injected state (gas-rich)
        Snew[inj] = 1.0 - Swr

    if phase == "withdraw":
        # Optional: mimic gas being removed at producer (keeps drawdown visible)
        # Comment this out if you want "natural" production only.
        Snew[prd] = Sgr

    return clampSg(Snew)
```

```python
# Initial gas saturation
Sg = np.ones((Nx, Ny)) * Sgr

# Well source term shape (point wells)
def make_q(q_inj, q_prd):
    q = np.zeros((Nx, Ny))
    q[inj] = q_inj
    q[prd] = q_prd
    q[0, 0] = 0.0  # keep reference cell clean
    return q

day = 24*3600

schedule = [
    ("inject",   6*day,  +5.0e-4,  -1.0e-6),  # inject at inj, produce at prd
    ("shutin",   2*day,   0.0,      0.0),
    ("withdraw", 6*day,  -1.0e-6,  +5.0e-4),  # reverse (toy withdrawal phase)
]

# Choose dt by CFL-like rule from initial pressure/velocity magnitude (auto-adjust)
def pick_dt(u_x, u_y, CFL=0.35):
    umax = max(np.max(np.abs(u_x)), np.max(np.abs(u_y)), 1e-12)
    hmin = min(dx, dy)
    return CFL * phi * hmin / umax

snapshots = []
t = 0.0

for phase, duration, qinj, qprd in schedule:
    q = make_q(qinj, qprd)
    p = solve_pressure(q)
    u_x, u_y = velocities_from_pressure(p, k, mu, dx, dy)

    dt = 0.1 * 24 * 3600   # 0.1 day timestep
    nsteps = int(np.ceil(duration/dt))

    for n in range(nsteps):
        Sg = transport_step(Sg, u_x, u_y, dt, phi, dx, dy, phase, inj, prd, Swr, Sgr)
        t += dt

        # store occasional snapshots for plotting
        if n % max(1, nsteps//6) == 0:
            snapshots.append((t, phase, p.copy(), Sg.copy()))

print("Snapshots stored:", len(snapshots))

# -----------------------------
# EXPORT DATASET PROPERLY
# -----------------------------

times = np.array([snap[0] for snap in snapshots])
phases = np.array([snap[1] for snap in snapshots])
P_all = np.stack([snap[2] for snap in snapshots])
Sg_all = np.stack([snap[3] for snap in snapshots])

# reconstruct well rates (must match your schedule)
phase_to_rates = {
    "inject":   (+5.0e-4, -5.0e-4),
    "shutin":   (0.0, 0.0),
    "withdraw": (-5.0e-4, +5.0e-4)
}

qinj = np.array([phase_to_rates[ph][0] for ph in phases])
qprd = np.array([phase_to_rates[ph][1] for ph in phases])

np.savez_compressed(
    "uhs_2d_dataset_fields.npz",
    times=times,
    phases=phases,
    qinj=qinj,
    qprd=qprd,
    P=P_all,
    Sg=Sg_all,
    Nx=Nx, Ny=Ny,
    Lx=Lx, Ly=Ly,
    dx=dx, dy=dy,
    Sgr=Sgr, Swr=Swr
)

print("Dataset saved correctly.")
```

```python
!pip -q install torch pandas numpy

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
```

```python
df = pd.read_csv("uhs_2d_dataset_pointwise.csv")

# --- constants (must match your simulator) ---
phi = 0.20
k   = 150e-15
mu  = 1.0e-3
mu_g = 1.0e-5
mu_w = 8.0e-4
Sgr, Swr = 0.05, 0.20

# --- normalize inputs (important for PINNs) ---
x_min, x_max = df["x"].min(), df["x"].max()
y_min, y_max = df["y"].min(), df["y"].max()
t_min, t_max = df["t_day"].min(), df["t_day"].max()

def norm_x(x): return 2*(x - x_min)/(x_max - x_min) - 1
def norm_y(y): return 2*(y - y_min)/(y_max - y_min) - 1
def norm_t(t): return 2*(t - t_min)/(t_max - t_min) - 1

# normalize pressure for stable training
p_mean, p_std = df["p"].mean(), df["p"].std()
def norm_p(p): return (p - p_mean)/p_std
def denorm_p(pn): return pn*p_std + p_mean

# normalize saturation to [0,1] effective saturation
def norm_S(S): return (S - Sgr) / (1.0 - Swr - Sgr)
def denorm_S(Sn): return Sn*(1.0 - Swr - Sgr) + Sgr
```

```python
class MLP(nn.Module):
    def __init__(self, in_dim, out_dim, width=128, depth=5):
        super().__init__()
        layers = [nn.Linear(in_dim, width), nn.Tanh()]
        for _ in range(depth-1):
            layers += [nn.Linear(width, width), nn.Tanh()]
        layers += [nn.Linear(width, out_dim)]
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)

# Inputs: x,y,t, phase, qinj, qprd  => 6 features
p_net = MLP(in_dim=6, out_dim=1, width=128, depth=5)
s_net = MLP(in_dim=6, out_dim=1, width=128, depth=5)

device = "cuda" if torch.cuda.is_available() else "cpu"
p_net.to(device); s_net.to(device)
```

```python
# Put your actual well coordinates here (meters)
# From your plots it looks like around (xâ‰ˆ250m,yâ‰ˆ500m) and (xâ‰ˆ750m,yâ‰ˆ500m) if Lx=Ly=1000
inj_xy = (250.0, 500.0)
prd_xy = (750.0, 500.0)

# Gaussian width (meters). Increase if training unstable
sigma = 40.0

def well_gaussian(x, y, x0, y0, sig):
    return torch.exp(-((x-x0)**2 + (y-y0)**2)/(2*sig**2))

def q_source(x, y, qinj, qprd):
    """
    Differentiable source term q(x,y) = qinj * G_inj - qprd * G_prd
    Note: sign convention depends on your pressure PDE; keep consistent with your simulator.
    """
    Gin = well_gaussian(x, y, inj_xy[0], inj_xy[1], sigma)
    Gpr = well_gaussian(x, y, prd_xy[0], prd_xy[1], sigma)
    return qinj*Gin + qprd*Gpr
```

```python
def fg_torch(Sg):
    # clamp Sg to physical bounds
    Sg = torch.clamp(Sg, Sgr, 1.0 - Swr)

    Se_g = (Sg - Sgr) / (1.0 - Swr - Sgr)
    Se_w = (1.0 - Sg - Swr) / (1.0 - Swr - Sgr)

    krg = Se_g**2.0
    krw = Se_w**2.0

    lg = krg / mu_g
    lw = krw / mu_w
    return lg/(lg+lw)
```

```python
# Convert df columns to numpy once
data_np = df[["x","y","t_day","phase","qinj","qprd","p","Sg"]].to_numpy()

def sample_data(n):
    idx = np.random.choice(len(data_np), size=n, replace=False)
    batch = data_np[idx]
    x,y,t,ph,qinj,qprd,p,Sg = batch.T
    X = np.stack([norm_x(x), norm_y(y), norm_t(t), ph, qinj, qprd], axis=1).astype(np.float32)
    y_p = norm_p(p).astype(np.float32)[:,None]
    y_s = norm_S(Sg).astype(np.float32)[:,None]
    return X, y_p, y_s

def sample_collocation(n):
    # uniform in domain/time using observed ranges
    x = np.random.uniform(x_min, x_max, size=n)
    y = np.random.uniform(y_min, y_max, size=n)
    t = np.random.uniform(t_min, t_max, size=n)

    # sample phases/rates from observed unique combos to stay realistic
    combos = df[["phase","qinj","qprd"]].drop_duplicates().to_numpy()
    cidx = np.random.choice(len(combos), size=n, replace=True)
    ph, qinj, qprd = combos[cidx].T

    X = np.stack([norm_x(x), norm_y(y), norm_t(t), ph, qinj, qprd], axis=1).astype(np.float32)
    return X
```

```python
def gradients(u, x):
    return torch.autograd.grad(u, x, grad_outputs=torch.ones_like(u), create_graph=True)[0]

def physics_losses(Xc_np):
    Xc = torch.tensor(Xc_np, device=device, requires_grad=True)

    # Split normalized inputs directly from Xc
    xn = Xc[:,0:1]
    yn = Xc[:,1:2]
    tn = Xc[:,2:3]
    ph = Xc[:,3:4]
    qinj = Xc[:,4:5]
    qprd = Xc[:,5:6]

    # Predict normalized pressure and saturation
    pn = p_net(Xc)
    Sn_eff = torch.sigmoid(s_net(Xc))

    # Convert to physical units
    p = denorm_p(pn)
    Sg = denorm_S(Sn_eff)

    # ----------------------------
    # GRADIENTS wrt Xc
    # ----------------------------
    grads_p = torch.autograd.grad(p, Xc,
                                  grad_outputs=torch.ones_like(p),
                                  create_graph=True)[0]

    dp_dx = grads_p[:,0:1]
    dp_dy = grads_p[:,1:2]

    grads_S = torch.autograd.grad(Sg, Xc,
                                  grad_outputs=torch.ones_like(Sg),
                                  create_graph=True)[0]

    dS_dt = grads_S[:,2:3]

    # Darcy velocity
    ux = -(k/mu) * dp_dx
    uy = -(k/mu) * dp_dy

    # Pressure Laplacian
    d2p_dx2 = torch.autograd.grad(dp_dx, Xc,
                                  grad_outputs=torch.ones_like(dp_dx),
                                  create_graph=True)[0][:,0:1]

    d2p_dy2 = torch.autograd.grad(dp_dy, Xc,
                                  grad_outputs=torch.ones_like(dp_dy),
                                  create_graph=True)[0][:,1:2]

    div_term = (k/mu) * (d2p_dx2 + d2p_dy2)

    # Convert normalized x,y back to meters for well source
    x_phys = (xn + 1)/2 * (x_max - x_min) + x_min
    y_phys = (yn + 1)/2 * (y_max - y_min) + y_min

    q = q_source(x_phys, y_phys, qinj, qprd)

    res_p = div_term - q
    loss_pde_p = torch.mean(res_p**2)

    # Saturation residual
    f = fg_torch(Sg)

    Fx = ux * f
    Fy = uy * f

    dFx_dx = torch.autograd.grad(Fx, Xc,
                                 grad_outputs=torch.ones_like(Fx),
                                 create_graph=True)[0][:,0:1]

    dFy_dy = torch.autograd.grad(Fy, Xc,
                                 grad_outputs=torch.ones_like(Fy),
                                 create_graph=True)[0][:,1:2]

    res_s = dS_dt + (1.0/phi) * (dFx_dx + dFy_dy)
    loss_pde_s = torch.mean(res_s**2)

    return loss_pde_p, loss_pde_s

```

```python
opt = torch.optim.Adam(list(p_net.parameters()) + list(s_net.parameters()), lr=1e-3)

# weights (tune later)
w_data_p = 1.0
w_data_s = 1.0
w_pde_p  = 1.0
w_pde_s  = 1.0

for step in range(2000):
    # --- data batch ---
    Xd, yp, ys = sample_data(n=4096)
    Xd = torch.tensor(Xd, device=device)
    yp = torch.tensor(yp, device=device)
    ys = torch.tensor(ys, device=device)

    pn_pred = p_net(Xd)
    Sn_eff_pred = torch.sigmoid(s_net(Xd))  # (0,1)

    loss_data_p = torch.mean((pn_pred - yp)**2)
    # front-weighted saturation loss
    w = 1.0 + 3.0 * (ys > 0.15).float()   # 0.15 threshold can be tuned
    loss_data_s = torch.mean(w * (Sn_eff_pred - ys)**2)


    # --- collocation batch for physics ---
    Xc = sample_collocation(n=2048)
    loss_pde_p, loss_pde_s = physics_losses(Xc)

    loss = (w_data_p*loss_data_p + w_data_s*loss_data_s +
            w_pde_p*loss_pde_p + w_pde_s*loss_pde_s)

    opt.zero_grad()
    loss.backward()
    opt.step()

    if step % 100 == 0:
        print(f"step {step:4d} | L={loss.item():.3e} "
              f"| data_p={loss_data_p.item():.2e} data_s={loss_data_s.item():.2e} "
              f"| pde_p={loss_pde_p.item():.2e} pde_s={loss_pde_s.item():.2e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import torch

# ---- Load dataset ----
data = np.load("uhs_2d_dataset_fields.npz")
times_sec = data["times"]
phases_str = data["phases"]
qinj_arr = data["qinj"]
qprd_arr = data["qprd"]
Sg_sim_all = data["Sg"]

Nx = int(data["Nx"]); Ny = int(data["Ny"])
dx = float(data["dx"]); dy = float(data["dy"])
Sgr = float(data["Sgr"]); Swr = float(data["Swr"])
day = 24*3600
times_day = times_sec/day

# ---- Grid ----
x = (np.arange(Nx)+0.5)*dx
y = (np.arange(Ny)+0.5)*dy
Xg, Yg = np.meshgrid(x, y, indexing="ij")

# ---- Phase encoding (must match training) ----
# Corrected: Use integer IDs as keys to match the phases_str array content.
phase_to_id = {1: 1.0, 0: 0.0, -1: -1.0}

device = next(s_net.parameters()).device
s_net.eval()

def predict_Sg_field(idx):
    t0 = float(times_day[idx])
    # Corrected: Directly use the integer as a key, as phases_str contains integer IDs.
    ph0 = float(phase_to_id[phases_str[idx]])
    qinj0 = float(qinj_arr[idx])
    qprd0 = float(qprd_arr[idx])

    X_in = np.stack([
        norm_x(Xg.ravel()),
        norm_y(Yg.ravel()),
        norm_t(np.full(Xg.size, t0)),
        np.full(Xg.size, ph0),
        np.full(Xg.size, qinj0),
        np.full(Xg.size, qprd0),
    ], axis=1).astype(np.float32)

    out = np.zeros((X_in.shape[0], 1), dtype=np.float32)
    bs = 8192
    with torch.no_grad():
        for i in range(0, X_in.shape[0], bs):
            xb = torch.tensor(X_in[i:i+bs], device=device)
            out[i:i+bs] = torch.sigmoid(s_net(xb)).cpu().numpy()

    return denorm_S(out).reshape(Nx, Ny)

# ---- Choose which snapshots to include ----
# Option A (fast): just a few snapshots
sel = [0, 5, 9, 13, len(times_day)-1]

# Option B (heavier): all snapshots
# sel = list(range(len(times_day)))

# ---- Sample points (so plot isn't too dense) ----
rng = np.random.default_rng(0)
n_points_per_snapshot = 2000   # increase if you want denser plot

y_true = []
y_pred = []

for idx in sel:
    Sg_pred = predict_Sg_field(idx)
    Sg_true = Sg_sim_all[idx]

    # random sample indices
    flatN = Nx*Ny
    pick = rng.choice(flatN, size=min(n_points_per_snapshot, flatN), replace=False)

    y_true.append(Sg_true.ravel()[pick])
    y_pred.append(Sg_pred.ravel()[pick])

y_true = np.concatenate(y_true)
y_pred = np.concatenate(y_pred)

# ---- Plot parity ----
plt.figure(figsize=(6,6))
plt.scatter(y_true, y_pred, s=12, alpha=0.35, marker='x')  # orange-style "x"
mn = min(y_true.min(), y_pred.min())
mx = max(y_true.max(), y_pred.max())
plt.plot([mn, mx], [mn, mx], 'k--', linewidth=2)          # dashed y=x
plt.xlabel("Simulator $S_g$")
plt.ylabel("PINN Predicted $S_g$")
plt.title("Parity Plot: PINN vs Simulator (Sg)")
plt.grid(True, alpha=0.25)
plt.tight_layout()
plt.show()

# ---- Metrics on sampled points ----
rmse = np.sqrt(np.mean((y_pred - y_true)**2))
mae  = np.mean(np.abs(y_pred - y_true))
print("Sampled RMSE:", rmse, " | Sampled MAE:", mae)
```
