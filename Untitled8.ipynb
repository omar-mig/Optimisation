````python
# --- Grid ---
Nx = 200
L  = 1000.0     # meters
dx = L / Nx
x  = np.linspace(dx/2, L-dx/2, Nx)

# --- Rock ---
phi = 0.20      # porosity (constant for baseline)
k   = 100e-15   # permeability in m^2 (â‰ˆ100 mD)

# --- Fluids (gas=H2, water=brine) ---
mu_g = 1.0e-5   # Pa.s (order of magnitude for H2)
mu_w = 8.0e-4   # Pa.s (brine-ish)

# --- Corey-type relperm params ---
Sgr = 0.05      # residual gas
Swr = 0.20      # residual water

ng = 2.0
nw = 2.0

def clampSg(Sg):
    return np.clip(Sg, Sgr, 1.0 - Swr)

def krg(Sg):
    Sg = clampSg(Sg)
    Se = (Sg - Sgr) / (1.0 - Swr - Sgr)
    return Se**ng

def krw(Sg):
    Sg = clampSg(Sg)
    Sw = 1.0 - Sg
    Se = (Sw - Swr) / (1.0 - Swr - Sgr)
    return Se**nw

def fractional_flow_g(Sg):
    lg = krg(Sg) / mu_g
    lw = krw(Sg) / mu_w
    return lg / (lg + lw)
````

````python
def step_transport(Sg, u, dt):
    """
    1D upwind scheme for: dS/dt + (u/phi) d f(S)/dx = 0
    """
    f = fractional_flow_g(Sg)
    flux = u * f  # volumetric flux of gas fraction

    # Upwind numerical flux
    F = np.zeros(Nx+1)
    if u >= 0:
        F[1:Nx] = flux[:-1]
        F[0]    = u * fractional_flow_g(1.0 - Swr)  # injected gas saturation state
        F[Nx]   = flux[-1]
    else:
        F[1:Nx] = flux[1:]
        F[0]    = flux[0]
        F[Nx]   = u * fractional_flow_g(Sgr)         # production boundary

    # Update
    dS = -(dt/(phi*dx)) * (F[1:] - F[:-1])
    Snew = clampSg(Sg + dS)
    return Snew
````

````python
# Initial condition: mostly brine (low gas)
Sg = np.ones(Nx) * Sgr

# Schedule (seconds)
day = 24*3600
schedule = [
    ("inject",   5*day,  +1.0e-5),  # u > 0 pushes gas in
    ("shutin",   2*day,   0.0),
    ("withdraw", 5*day,  -1.0e-5),  # u < 0 pulls gas back
]

dt = 0.2*day  # time step; reduce if unstable

history = []
time = 0.0

for (name, duration, u) in schedule:
    nsteps = int(np.ceil(duration/dt))
    for _ in range(nsteps):
        Sg = step_transport(Sg, u, dt)
        time += dt
        history.append((time, name, Sg.copy()))
````

````python
# ---------------------------
# Grid
# ---------------------------
Nx, Ny = 80, 80
Lx, Ly = 1000.0, 1000.0  # meters
dx, dy = Lx/Nx, Ly/Ny

x = (np.arange(Nx) + 0.5) * dx
y = (np.arange(Ny) + 0.5) * dy
X, Y = np.meshgrid(x, y, indexing="ij")

# ---------------------------
# Rock/fluid (baseline)
# ---------------------------
phi = 0.20
k   = 150e-15      # m^2 (â‰ˆ150 mD)
mu  = 1.0e-3       # Pa.s (use single-phase viscosity for pressure solve baseline)

# ---------------------------
# Relperm / fractional flow for transport
# ---------------------------
mu_g = 1.0e-5
mu_w = 8.0e-4
Sgr, Swr = 0.05, 0.20
ng, nw = 2.0, 2.0

def clampSg(Sg):
    return np.clip(Sg, Sgr, 1.0 - Swr)

def krg(Sg):
    Sg = clampSg(Sg)
    Se = (Sg - Sgr) / (1.0 - Swr - Sgr)
    return Se**ng

def krw(Sg):
    Sg = clampSg(Sg)
    Sw = 1.0 - Sg
    Se = (Sw - Swr) / (1.0 - Swr - Sgr)
    return Se**nw

def fg(Sg):
    lg = krg(Sg)/mu_g
    lw = krw(Sg)/mu_w
    return lg/(lg+lw)
````

````python
def ij_to_id(i, j):
    return i*Ny + j

def build_pressure_matrix(Nx, Ny, dx, dy, k, mu):
    N = Nx*Ny
    A = lil_matrix((N, N))
    T_x = (k/mu) / dx**2
    T_y = (k/mu) / dy**2

    for i in range(Nx):
        for j in range(Ny):
            pid = ij_to_id(i, j)
            diag = 0.0

            # West/East
            if i > 0:
                nid = ij_to_id(i-1, j)
                A[pid, nid] = -T_x
                diag += T_x
            if i < Nx-1:
                nid = ij_to_id(i+1, j)
                A[pid, nid] = -T_x
                diag += T_x

            # South/North
            if j > 0:
                nid = ij_to_id(i, j-1)
                A[pid, nid] = -T_y
                diag += T_y
            if j < Ny-1:
                nid = ij_to_id(i, j+1)
                A[pid, nid] = -T_y
                diag += T_y

            A[pid, pid] = diag

    # Fix gauge (pressure reference) at cell (0,0)
    ref = ij_to_id(0, 0)
    A[ref, :] = 0.0
    A[ref, ref] = 1.0
    return A.tocsr()

A = build_pressure_matrix(Nx, Ny, dx, dy, k, mu)

def solve_pressure(q):
    """
    q is Nx x Ny source term (positive injection, negative production).
    """
    b = q.reshape(-1).copy()
    b[ij_to_id(0, 0)] = 0.0  # reference pressure cell
    p = spsolve(A, b).reshape((Nx, Ny))
    return p
````

````python
inj = (int(0.25*Nx), int(0.50*Ny))
prd = (int(0.75*Nx), int(0.50*Ny))

print("Injector cell:", inj, "Producer cell:", prd)
````

````python
def velocities_from_pressure(p, k, mu, dx, dy):
    """
    Returns face velocities u_x (Nx+1, Ny), u_y (Nx, Ny+1).
    Positive u_x is +x direction; positive u_y is +y direction.
    """
    Tx = (k/mu)/dx
    Ty = (k/mu)/dy

    u_x = np.zeros((Nx+1, Ny))
    u_y = np.zeros((Nx, Ny+1))

    # Internal faces: u = - (k/mu) dp/dx
    # x-faces between i-1 and i: use p[i]-p[i-1]
    for i in range(1, Nx):
        u_x[i, :] = -Tx * (p[i, :] - p[i-1, :])

    for j in range(1, Ny):
        u_y[:, j] = -Ty * (p[:, j] - p[:, j-1])

    # No-flow boundaries already enforced by leaving boundary faces = 0
    return u_x, u_y
````

````python
def transport_step(Sg, u_x, u_y, dt, phi, dx, dy, phase, inj, prd, Swr, Sgr):
    """
    Conservative upwind transport:
      dS/dt + div( u * f(S) ) / phi = 0

    Added well coupling:
    - During injection: enforce high gas saturation in injector cell (Dirichlet).
    - Optionally during withdrawal: enforce low gas saturation at producer to mimic removal.
    """
    Sg = clampSg(Sg)
    f = fg(Sg)

    Fx = np.zeros((Nx+1, Ny))
    Fy = np.zeros((Nx, Ny+1))

    # x-faces (upwind)
    for i in range(1, Nx):
        ux = u_x[i, :]
        up = (ux >= 0)
        fL = f[i-1, :]
        fR = f[i, :]
        Fx[i, :] = ux * np.where(up, fL, fR)

    # y-faces (upwind)
    for j in range(1, Ny):
        uy = u_y[:, j]
        up = (uy >= 0)
        fB = f[:, j-1]
        fT = f[:, j]
        Fy[:, j] = uy * np.where(up, fB, fT)

    divF = (Fx[1:, :] - Fx[:-1, :]) / dx + (Fy[:, 1:] - Fy[:, :-1]) / dy
    Snew = clampSg(Sg - (dt/phi) * divF)

    # -----------------------------
    # WELL COUPLING (THE FIX)
    # -----------------------------
    if phase == "inject":
        # Force injector cell to injected state (gas-rich)
        Snew[inj] = 1.0 - Swr

    if phase == "withdraw":
        # Optional: mimic gas being removed at producer (keeps drawdown visible)
        # Comment this out if you want "natural" production only.
        Snew[prd] = Sgr

    return clampSg(Snew)
````

````python
# Initial gas saturation
Sg = np.ones((Nx, Ny)) * Sgr

# Well source term shape (point wells)
def make_q(q_inj, q_prd):
    q = np.zeros((Nx, Ny))
    q[inj] = q_inj
    q[prd] = q_prd
    q[0, 0] = 0.0  # keep reference cell clean
    return q

day = 24*3600

schedule = [
    ("inject",   6*day,  +5.0e-4,  -1.0e-6),  # inject at inj, produce at prd
    ("shutin",   2*day,   0.0,      0.0),
    ("withdraw", 6*day,  -1.0e-6,  +5.0e-4),  # reverse (toy withdrawal phase)
]

# Choose dt by CFL-like rule from initial pressure/velocity magnitude (auto-adjust)
def pick_dt(u_x, u_y, CFL=0.35):
    umax = max(np.max(np.abs(u_x)), np.max(np.abs(u_y)), 1e-12)
    hmin = min(dx, dy)
    return CFL * phi * hmin / umax

snapshots = []
t = 0.0

for phase, duration, qinj, qprd in schedule:
    q = make_q(qinj, qprd)
    p = solve_pressure(q)
    u_x, u_y = velocities_from_pressure(p, k, mu, dx, dy)

    dt = 0.1 * 24 * 3600   # 0.1 day timestep
    nsteps = int(np.ceil(duration/dt))

    for n in range(nsteps):
        Sg = transport_step(Sg, u_x, u_y, dt, phi, dx, dy, phase, inj, prd, Swr, Sgr)
        t += dt

        # store occasional snapshots for plotting
        if n % max(1, nsteps//6) == 0:
            snapshots.append((t, phase, p.copy(), Sg.copy()))

print("Snapshots stored:", len(snapshots))

# -----------------------------
# EXPORT DATASET PROPERLY
# -----------------------------

times = np.array([snap[0] for snap in snapshots])
phases = np.array([snap[1] for snap in snapshots])
P_all = np.stack([snap[2] for snap in snapshots])
Sg_all = np.stack([snap[3] for snap in snapshots])

# reconstruct well rates (must match your schedule)
phase_to_rates = {
    "inject":   (+5.0e-4, -5.0e-4),
    "shutin":   (0.0, 0.0),
    "withdraw": (-5.0e-4, +5.0e-4)
}

qinj = np.array([phase_to_rates[ph][0] for ph in phases])
qprd = np.array([phase_to_rates[ph][1] for ph in phases])

np.savez_compressed(
    "uhs_2d_dataset_fields.npz",
    times=times,
    phases=phases,
    qinj=qinj,
    qprd=qprd,
    P=P_all,
    Sg=Sg_all,
    Nx=Nx, Ny=Ny,
    Lx=Lx, Ly=Ly,
    dx=dx, dy=dy,
    Sgr=Sgr, Swr=Swr
)

print("Dataset saved correctly.")
````

````python
class MLP(nn.Module):
    def __init__(self, in_dim, out_dim, width=128, depth=5):
        super().__init__()
        layers = [nn.Linear(in_dim, width), nn.Tanh()]
        for _ in range(depth-1):
            layers += [nn.Linear(width, width), nn.Tanh()]
        layers += [nn.Linear(width, out_dim)]
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)

# Inputs: x,y,t, phase, qinj, qprd  => 6 features
p_net = MLP(in_dim=6, out_dim=1, width=128, depth=5)
s_net = MLP(in_dim=6, out_dim=1, width=128, depth=5)

device = "cuda" if torch.cuda.is_available() else "cpu"
p_net.to(device); s_net.to(device)
````

````python
def gradients(u, x):
    return torch.autograd.grad(u, x, grad_outputs=torch.ones_like(u), create_graph=True)[0]

def physics_losses(Xc_np):
    Xc = torch.tensor(Xc_np, device=device, requires_grad=True)

    # Split normalized inputs directly from Xc
    xn = Xc[:,0:1]
    yn = Xc[:,1:2]
    tn = Xc[:,2:3]
    ph = Xc[:,3:4]
    qinj = Xc[:,4:5]
    qprd = Xc[:,5:6]

    # Predict normalized pressure and saturation
    pn = p_net(Xc)
    Sn_eff = torch.sigmoid(s_net(Xc))

    # Convert to physical units
    p = denorm_p(pn)
    Sg = denorm_S(Sn_eff)

    # ----------------------------
    # GRADIENTS wrt Xc
    # ----------------------------
    grads_p = torch.autograd.grad(p, Xc,
                                  grad_outputs=torch.ones_like(p),
                                  create_graph=True)[0]

    dp_dx = grads_p[:,0:1]
    dp_dy = grads_p[:,1:2]

    grads_S = torch.autograd.grad(Sg, Xc,
                                  grad_outputs=torch.ones_like(Sg),
                                  create_graph=True)[0]

    dS_dt = grads_S[:,2:3]

    # Darcy velocity
    ux = -(k/mu) * dp_dx
    uy = -(k/mu) * dp_dy

    # Pressure Laplacian
    d2p_dx2 = torch.autograd.grad(dp_dx, Xc,
                                  grad_outputs=torch.ones_like(dp_dx),
                                  create_graph=True)[0][:,0:1]

    d2p_dy2 = torch.autograd.grad(dp_dy, Xc,
                                  grad_outputs=torch.ones_like(dp_dy),
                                  create_graph=True)[0][:,1:2]

    div_term = (k/mu) * (d2p_dx2 + d2p_dy2)

    # Convert normalized x,y back to meters for well source
    x_phys = (xn + 1)/2 * (x_max - x_min) + x_min
    y_phys = (yn + 1)/2 * (y_max - y_min) + y_min

    q = q_source(x_phys, y_phys, qinj, qprd)

    res_p = div_term - q
    loss_pde_p = torch.mean(res_p**2)

    # Saturation residual
    f = fg_torch(Sg)

    Fx = ux * f
    Fy = uy * f

    dFx_dx = torch.autograd.grad(Fx, Xc,
                                 grad_outputs=torch.ones_like(Fx),
                                 create_graph=True)[0][:,0:1]

    dFy_dy = torch.autograd.grad(Fy, Xc,
                                 grad_outputs=torch.ones_like(Fy),
                                 create_graph=True)[0][:,1:2]

    res_s = dS_dt + (1.0/phi) * (dFx_dx + dFy_dy)
    loss_pde_s = torch.mean(res_s**2)

    return loss_pde_p, loss_pde_s

````

````python
opt = torch.optim.Adam(list(p_net.parameters()) + list(s_net.parameters()), lr=1e-3)

# weights (tune later)
w_data_p = 1.0
w_data_s = 1.0
w_pde_p  = 1.0
w_pde_s  = 1.0

# ---- loss history ----
steps_hist = []
train_hist = []
test_hist  = []
log_every = 50

def sample_collocation(n):
    # Randomly sample within the normalized ranges for x, y, t
    xc = np.random.uniform(-1, 1, size=(n, 1))
    yc = np.random.uniform(-1, 1, size=(n, 1))
    tc = np.random.uniform(-1, 1, size=(n, 1))

    # Randomly sample phase and rates from the unique combinations in the dataset
    # This assumes 'df' is available and correctly loaded from uhs_2d_dataset_pointwise.csv
    combos = df[["phase", "qinj", "qprd"]].drop_duplicates().to_numpy()
    cidx = np.random.choice(len(combos), size=n, replace=True)
    phc, qinjc, qprdc = combos[cidx].T.reshape(3, n, 1)

    Xc = np.concatenate([xc, yc, tc, phc, qinjc, qprdc], axis=1).astype(np.float32)
    return Xc

for step in range(2000):
    # --- TRAIN data batch ---
    Xd, yp, ys = sample_train_data(n=4096)
    Xd = torch.tensor(Xd, device=device).float()
    yp = torch.tensor(yp, device=device).float()
    ys = torch.tensor(ys, device=device).float()

    pn_pred = p_net(Xd)
    Sn_eff_pred = torch.sigmoid(s_net(Xd))  # (0,1)

    loss_data_p = torch.mean((pn_pred - yp)**2)

    # front-weighted saturation loss (train)
    w = 1.0 + 3.0 * (ys > 0.15).float()
    loss_data_s = torch.mean(w * (Sn_eff_pred - ys)**2)

    # --- collocation batch for physics ---
    Xc = sample_collocation(n=2048)
    loss_pde_p, loss_pde_s = physics_losses(Xc)

    loss = (w_data_p*loss_data_p + w_data_s*loss_data_s +
            w_pde_p*loss_pde_p + w_pde_s*loss_pde_s)

    opt.zero_grad()
    loss.backward()
    opt.step()

    # ---- LOG train + test losses ----
    if step % log_every == 0:
        # training loss (full)
        train_loss_val = loss.item()

        # test loss (data only)
        Xd_te, yp_te, ys_te = sample_test_data(n=4096)
        Xd_te = torch.tensor(Xd_te, device=device).float()
        yp_te = torch.tensor(yp_te, device=device).float()
        ys_te = torch.tensor(ys_te, device=device).float()

        with torch.no_grad():
            pn_te = p_net(Xd_te)
            Sn_te = torch.sigmoid(s_net(Xd_te))

            test_data_p = torch.mean((pn_te - yp_te)**2)
            w_te = 1.0 + 3.0 * (ys_te > 0.15).float()
            test_data_s = torch.mean(w_te * (Sn_te - ys_te)**2)

            test_loss_val = (w_data_p*test_data_p + w_data_s*test_data_s).item()

        steps_hist.append(step)
        train_hist.append(train_loss_val)
        test_hist.append(test_loss_val)

    if step % 100 == 0:
        print(f"step {step:4d} | L={loss.item():.3e} "
              f"| data_p={loss_data_p.item():.2e} data_s={loss_data_s.item():.2e} "
              f"| pde_p={loss_pde_p.item():.2e} pde_s={loss_pde_s.item():.2e}")
````

````python
import torch
import numpy as np

def run_pinn_forward(qinj_opt, qprd_opt, current_time_day, current_phase_id_int):
    """
    Runs the PINN forward for a single time step with optimized rates.
    Returns predicted pressure and saturation fields.
    """
    # Ensure qinj_opt and qprd_opt are tensors for gradient tracking
    # These tensors are passed directly to the PINN input construction.
    qinj_tensor = qinj_opt.clone().detach().requires_grad_(True) # Ensure it requires grad
    qprd_tensor = qprd_opt.clone().detach().requires_grad_(True) # Ensure it requires grad

    # Convert phase_id to float for the prediction functions
    # The predict_ functions expect `phase_str` as the argument name, but it's an int here.
    # Let's map it back to the string key that phase_to_id expects (for simplicity with existing functions)
    phase_str_map = {1: "inject", 0: "shutin", -1: "withdraw"}
    actual_phase_str = phase_str_map[current_phase_id_int]

    # Construct the input tensor for the network, ensuring qinj_tensor and qprd_tensor are part of the graph
    # We need to manually build the input tensor that the PINN expects for *all* grid points
    # This is effectively what predict_P_field/predict_Sg_field do internally, but we need to do it
    # in a way that passes the differentiable qinj_tensor/qprd_tensor.

    # Create a full grid of input features for the PINN
    # Xg_t and Yg_t are already torch tensors defined globally
    t_day_t = torch.tensor(current_time_day, dtype=torch.float32, device=device)
    ph0_t = torch.tensor(float(phase_to_id[current_phase_id_int]), dtype=torch.float32, device=device)

    # Expand qinj_tensor and qprd_tensor to match the size of Xg_t.ravel()
    qinj_expanded = torch.full_like(Xg_t.ravel(), qinj_tensor.item()) # Use .item() then full_like to expand
    qprd_expanded = torch.full_like(Xg_t.ravel(), qprd_tensor.item()) # This is a conceptual simplification

    # For true differentiability, you might need to pass qinj_tensor and qprd_tensor as variables
    # in a more direct way within the PINN input construction, for instance:
    X_in_t = torch.stack([
        norm_x_t(Xg_t.ravel()),
        norm_y_t(Yg_t.ravel()),
        norm_t_t(torch.full_like(Xg_t.ravel(), t_day_t)),
        torch.full_like(Xg_t.ravel(), ph0_t),
        qinj_expanded, # Now this part should be linked to the original qinj_tensor's grad
        qprd_expanded, # And this to qprd_tensor's grad
    ], axis=1)

    # --- Actual Differentiable PINN Forward Pass ---
    # Since the global predict_P_field/predict_Sg_field return numpy, we must redo the call
    # This ensures the computation remains within the torch graph for gradient calculation

    pn_pred = p_net(X_in_t) # Normalized pressure prediction
    Sn_eff_pred = torch.sigmoid(s_net(X_in_t)) # Normalized effective saturation prediction

    P_pred_tensor = denorm_p_t(pn_pred).reshape(Nx, Ny)
    Sg_pred_tensor = denorm_S_t(Sn_eff_pred).reshape(Nx, Ny)

    # Returning the original qinj_opt/qprd_opt (which are the ones requiring grad) for the objective
    return P_pred_tensor, Sg_pred_tensor, qinj_opt, qprd_opt


def objective_function(P_pred, Sg_pred, qinj, qprd):
    """
    Example objective: Maximize average Sg while minimizing injection rates.
    Note: This is a placeholder. Your actual objective will depend on your goals.
    """
    avg_Sg = Sg_pred.mean()
    # Assume a cost for injection/production
    injection_cost = qinj * 1e6 # example coefficient
    production_value = qprd * 1e6 # example coefficient

    # We want to maximize Sg and (production - injection cost)
    # So we minimize - (avg_Sg + production_value - injection_cost)
    # We also want to keep P within bounds, for simplicity, let's add a penalty
    target_P_max = 1.5e9
    P_penalty = torch.relu(P_pred - target_P_max).mean() * 1e-10 # Penalize exceeding max pressure

    # Return the value to be minimized
    return - (avg_Sg + production_value - injection_cost) + P_penalty


# --- Optimization Setup ---

# Current state (example values)
current_time_day = 6.0 # Example: mid-shutin
current_phase_id_int = 0 # 0 for shutin (shutin maps to 0.0 in phase_to_id)

# Initialize optimization variables (injection/production rates)
# These need to be torch tensors that require gradients
initial_qinj = torch.tensor(1e-5, requires_grad=True, device=device) # Example non-zero initial guess
initial_qprd = torch.tensor(-1e-5, requires_grad=True, device=device) # Example non-zero initial guess

# Define the optimizer for the rates
optimizer_rates = torch.optim.Adam([initial_qinj, initial_qprd], lr=1e-7) # Adjusted LR for stability

# Number of optimization steps
n_opt_steps = 100

print("Starting optimization...")
for opt_step in range(n_opt_steps):
    optimizer_rates.zero_grad()

    # Forward pass the PINN with current rates
    P_pred, Sg_pred, qinj_current, qprd_current = run_pinn_forward(
        initial_qinj, initial_qprd, current_time_day, current_phase_id_int
    )

    # Calculate the objective function
    loss_objective = objective_function(P_pred, Sg_pred, qinj_current, qprd_current)

    # Backward pass to compute gradients
    loss_objective.backward() # Computes gradients of loss_objective w.r.t initial_qinj, initial_qprd

    # Update the injection/production rates
    optimizer_rates.step()

    # Optional: Apply physical constraints (e.g., rates cannot be negative)
    with torch.no_grad():
        initial_qinj.clamp_(min=0.0) # Injection rates usually >= 0
        initial_qprd.clamp_(max=0.0) # Production rates usually <= 0 (negative flow)

    if opt_step % 10 == 0:
        print(f"Opt Step {opt_step:3d} | Objective: {loss_objective.item():.4f} | "
              f"Qinj: {initial_qinj.item():.2e} | Qprd: {initial_qprd.item():.2e}")

print("\nOptimization finished.")
print(f"Optimal Injection Rate (qinj): {initial_qinj.item():.2e}")
print(f"Optimal Production Rate (qprd): {initial_qprd.item():.2e}")
````

````python
opt_nn = torch.optim.Adam(list(p_net_nn.parameters()) + list(s_net_nn.parameters()), lr=1e-3)

train_hist_nn = []
test_hist_nn = []
steps_hist_nn = []

print("Starting training for data-driven neural networks...")

for step in range(2000):
    # --- TRAIN data batch ---
    Xd, yp, ys = sample_train_data(n=4096)
    Xd = torch.tensor(Xd, device=device).float()
    yp = torch.tensor(yp, device=device).float()
    ys = torch.tensor(ys, device=device).float()

    pn_pred_nn = p_net_nn(Xd)
    Sn_eff_pred_nn = torch.sigmoid(s_net_nn(Xd)) # (0,1) for effective saturation

    loss_data_p_nn = torch.mean((pn_pred_nn - yp)**2)
    # Optionally use front-weighted saturation loss for data-driven model too
    w_nn = 1.0 + 3.0 * (ys > 0.15).float() # Re-using the weighting from PINN training
    loss_data_s_nn = torch.mean(w_nn * (Sn_eff_pred_nn - ys)**2)

    loss_nn = loss_data_p_nn + loss_data_s_nn

    opt_nn.zero_grad()
    loss_nn.backward()
    opt_nn.step()

    # ---- LOG train + test losses ----
    if step % log_every == 0:
        # training loss (full)
        train_loss_val_nn = loss_nn.item()

        # test loss (data only)
        Xd_te, yp_te, ys_te = sample_test_data(n=4096)
        Xd_te = torch.tensor(Xd_te, device=device).float()
        yp_te = torch.tensor(yp_te, device=device).float()
        ys_te = torch.tensor(ys_te, device=device).float()

        with torch.no_grad():
            pn_te_nn = p_net_nn(Xd_te)
            Sn_te_nn = torch.sigmoid(s_net_nn(Xd_te))

            test_data_p_nn = torch.mean((pn_te_nn - yp_te)**2)
            w_te_nn = 1.0 + 3.0 * (ys_te > 0.15).float()
            test_data_s_nn = torch.mean(w_te_nn * (Sn_te_nn - ys_te)**2)

            test_loss_val_nn = (test_data_p_nn + test_data_s_nn).item()

        steps_hist_nn.append(step)
        train_hist_nn.append(train_loss_val_nn)
        test_hist_nn.append(test_loss_val_nn)

    if step % 100 == 0:
        print(f"NN Step {step:4d} | L_nn={loss_nn.item():.3e} "
              f"| data_p_nn={loss_data_p_nn.item():.2e} data_s_nn={loss_data_s_nn.item():.2e}")

print("Training for data-driven neural networks finished.")
````

## Define Prediction Functions for Data-Driven NN

### Subtask:
Create two new functions, `predict_Sg_NN(t_day, phase_str, qinj0, qprd0)` and `predict_P_NN(t_day, phase_str, qinj0, qprd0)`, that use the trained data-driven `s_net_nn` and `p_net_nn` models, respectively, to make predictions across the 2D grid. These functions will be similar to the `predict_Sg_field` and `predict_P_field` for the PINN, using tensor-aware normalization and returning numpy arrays for compatibility with plotting.


**Reasoning**:
The subtask requires defining two prediction functions, `predict_Sg_NN` and `predict_P_NN`, for the data-driven neural networks. These functions will take input parameters, convert them to normalized tensor format, use the respective trained models (`s_net_nn` and `p_net_nn`) to make predictions, and then denormalize and reshape the output to numpy arrays, similar to the existing PINN prediction functions.



````python
def predict_Sg_NN(t_day, phase_id_int, qinj0, qprd0):
    # Ensure s_net_nn is in evaluation mode
    s_net_nn.eval()

    # Convert inputs to torch tensors
    t_day_t = torch.tensor(t_day, dtype=torch.float32, device=device)
    ph0_t = torch.tensor(float(phase_to_id[phase_id_int]), dtype=torch.float32, device=device)
    qinj0_t = torch.tensor(qinj0, dtype=torch.float32, device=device)
    qprd0_t = torch.tensor(qprd0, dtype=torch.float32, device=device)

    # Construct input tensor for the network using tensor-aware normalization
    # Xg_t and Yg_t are assumed to be globally defined torch tensors for the grid
    X_in_t = torch.stack([
        norm_x_t(Xg_t.ravel()),
        norm_y_t(Yg_t.ravel()),
        norm_t_t(torch.full_like(Xg_t.ravel(), t_day_t)),
        torch.full_like(Xg_t.ravel(), ph0_t),
        torch.full_like(Xg_t.ravel(), qinj0_t),
        torch.full_like(Xg_t.ravel(), qprd0_t),
    ], axis=1)

    # Predict normalized effective saturation
    with torch.no_grad(): # Predictions for visualization don't need gradients
        Sn_eff_pred_t = torch.sigmoid(s_net_nn(X_in_t)) # Apply sigmoid for (0,1) range

    # Denormalize and reshape output
    Sg_pred_t = denorm_S_t(Sn_eff_pred_t).reshape(Nx, Ny)
    return Sg_pred_t.cpu().numpy() # Return numpy array

def predict_P_NN(t_day, phase_id_int, qinj0, qprd0):
    # Ensure p_net_nn is in evaluation mode
    p_net_nn.eval()

    # Convert inputs to torch tensors
    t_day_t = torch.tensor(t_day, dtype=torch.float32, device=device)
    ph0_t = torch.tensor(float(phase_to_id[phase_id_int]), dtype=torch.float32, device=device)
    qinj0_t = torch.tensor(qinj0, dtype=torch.float32, device=device)
    qprd0_t = torch.tensor(qprd0, dtype=torch.float32, device=device)

    # Construct input tensor for the network using tensor-aware normalization
    X_in_t = torch.stack([
        norm_x_t(Xg_t.ravel()),
        norm_y_t(Yg_t.ravel()),
        norm_t_t(torch.full_like(Xg_t.ravel(), t_day_t)),
        torch.full_like(Xg_t.ravel(), ph0_t),
        torch.full_like(Xg_t.ravel(), qinj0_t),
        torch.full_like(Xg_t.ravel(), qprd0_t),
    ], axis=1)

    # Predict normalized pressure
    with torch.no_grad(): # Predictions for visualization don't need gradients
        pn_pred_t = p_net_nn(X_in_t)

    # Denormalize and reshape output
    P_pred_t = denorm_p_t(pn_pred_t).reshape(Nx, Ny)
    return P_pred_t.cpu().numpy() # Return numpy array

print("Prediction functions `predict_Sg_NN` and `predict_P_NN` defined.")
````
